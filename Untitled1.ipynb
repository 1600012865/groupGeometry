```python
import matplotlib.pyplot as plt
from sagan_models import Generator
import torch
import os
import sys
import torch.nn as nn
from metric import get_metric
import numpy as np
z = 32
num = 10000
path = '/data-174/xuanjc/geometry/models/sagan_rec0/'
glist = os.listdir(path)
glist = [i for i in glist if 'G' in i]
glist.sort()
net = Generator(z_dim=z)
net.cuda()
net = nn.DataParallel(net)
res = []

def takesecond(ele):
    return ele[1]

for idx,i in enumerate(glist):
    
    w = torch.load(os.path.join(path,i))
    try:
        net.load_state_dict(w)
    except:
        print('degraded')
        continue
    else:
        pass
    sample = torch.randn(num,z,1,1)
    sample = net(sample)
    sample = sample/2 + 0.5
    loss = []
    for j in range(num):
      tmp = sample[j,0].detach().cpu().numpy()
      rr = get_metric(tmp)
      loss.append(rr[0])
    loss = np.average(np.array(loss))
    res.append((i,loss))
    print('%d/%d: %.3f'%(idx,len(glist),loss))
    
res.sort(key=takesecond)
print(res)
torch.save(res,'metric_average.pkl')
```
