```python
import matplotlib.pyplot as plt
from sagan_models import Generator
import torch
import os
import sys
from metric import get_metric
import numpy as np
import torch.nn as nn
import pandas as pd
z = 32
idx = str(sys.argv[1])
w = torch.load('/data-174/xuanjc/geometry/models/sagan_rec0/18240_G.pth')
csv = pd.read_csv('/data-174/xuanjc/geometry/csv/gaussian_rec_n0_conc32/rec.csv')
res = np.zeros([32,32])
'''
x = list(csv['cx'])
y = list(csv['cy'])
x = [i*32 for i in x]
y = [j*32 for j in y]
for i in range(len(csv)):
  res[int(x[i])][int(y[i])]+=1
'''
net = Generator(z_dim=z)
net.cuda()
net = nn.DataParallel(net)
net.load_state_dict(w)
num = 10000
sample = torch.randn(num,z,1,1)
sample = net(sample)
sample = sample/2 + 0.5
sample *= 255
corx = []
cory = []
for i in range(num):
  tmp = sample[i,0].detach().cpu().numpy()
  loss = get_metric(tmp)
  corx.append(loss[4])
  cory.append(loss[5])

plt.scatter(corx,cory,alpha=0.05)
plt.show()
```
