```python
import matplotlib.pyplot as plt
from sagan_models import Generator
import torch
import os
import sys
from metric import get_metric
import numpy as np
import torch.nn as nn

z = 32
idx = str(sys.argv[1])
w = torch.load('/data-174/xuanjc/geometry/models/sagan_rec0/18240_G.pth')
net = Generator(z_dim=z)
net.cuda()
net = nn.DataParallel(net)
net.load_state_dict(w)
num = 10000
sample = torch.randn(num,z,1,1)
sample = net(sample)
sample = sample/2 + 0.5
sample *= 255
plt.gray()
loss = 0
corx = []
cory = []
for i in range(num):
  tmp = sample[i,0].detach().cpu().numpy()
  res = get_metric(tmp)
  corx.append(res[2])
  cory.append(res[3])
  loss += res[0]
print(loss/num)
plt.scatter(corx,cory,alpha=0.01)
```
